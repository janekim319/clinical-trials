```python
import os
import xml.etree.ElementTree as ET

for filename in os.listdir("/content/unzipped"):
  print("Currently on: ", filename)
  newname = filename[:-4]
  tree = ET.parse("/content/unzipped/" +str(filename))
        
        
  root = tree.findall("./eligibility")
        
  if root == []:
      inclusion_file = newname + "_inclusion.txt"
      f = open(inclusion_file, "x")
            
      f.write('\n' + "gender:")
            
      if tree.find(".//gender") == None:
          f.write("N/A")
      else:
          f.write(tree.find(".//gender").text)
            
      f.write('\n' + "minimum age:")
      if tree.find(".//minimum_age") == None:
          f.write("N/A")
      else:
          f.write(tree.find(".//minimum_age").text)                
    
      f.write('\n' + "maximum age:")
        
      if tree.find(".//maximum_age") == None:
          f.write("N/A")
      else:
          f.write(tree.find(".//maximum_age").text)
                
      f.write('\n' + "healthy volunteers:")
            
      if tree.find(".//healthy_volunteers")== None:
          f.write("N/A")
      else:
          f.write(tree.find(".//healthy_volunteers").text)
            
      f.close()
        
            
      exclusion_file = newname + "_exclusion.txt"
      f = open(exclusion_file, "x")
        
            
            
      f.write("N/A")
      f.close()
            
            
  else:    
      eligibility_criteria = ""
      for eligibility in root:
          for textblock in eligibility.findall('.//textblock'):
              eligibility_criteria = eligibility_criteria + textblock.text
            
      inclusion_file = newname + "_inclusion.txt"
      f = open(inclusion_file, "x") 
             
        
      idx = eligibility_criteria.find('Exclusion Criteria')
            
      f.write(eligibility_criteria[:idx])
            
      f.write('\n' + "gender:")
            
      if tree.find(".//gender") == None:
          f.write("N/A")
      else:
          f.write(tree.find(".//gender").text)
            
      f.write('\n' + "minimum age:")
      if tree.find(".//minimum_age") == None:
          f.write("N/A")
      else:
          f.write(tree.find(".//minimum_age").text)                
    
      f.write('\n' + "maximum age:")
        
      if tree.find(".//maximum_age") == None:
          f.write("N/A")
      else:
          f.write(tree.find(".//maximum_age").text)
                
      f.write('\n' + "healthy volunteers:")
            
      if tree.find(".//healthy_volunteers")== None:
          f.write("N/A")
      else:
          f.write(tree.find(".//healthy_volunteers").text)
            
      f.close()
            
            
      exclusion_file = newname + "_exclusion.txt"
      f = open(exclusion_file, "x")
        
            
      if idx == -1:
          f.write("N/A")
      else:
          f.write(eligibility_criteria[idx:])
            
      f.close()
        
        
        
```
